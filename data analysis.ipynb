```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

offsetAccX=344.0
offsetAccY=493.5
offsetAccZ=434.0
gainAccX=4088.0
gainAccY=4096.5
gainAccZ=4135.0
offsetGyroX=(-57)
offsetGyroY=(-60)
offsetGyroZ=88
scaleGyroX=0.0558
scaleGyroY=0.0587
scaleGyroZ=0.0554
```

```python
def plot_data(filename):
    names=["accX", "accY", "accZ", "gyroX", "gyroY", "gyroZ"]
    data = pd.read_table(filename, sep=' ', names=names, header=None)
    actual = data.copy()
    actual["accX"] = (actual["accX"]-offsetAccX) / gainAccX
    actual["accY"] = (actual["accY"]-offsetAccY) / gainAccY
    actual["accZ"] = (actual["accZ"]-offsetAccZ) / gainAccZ
    actual["gyroX"] = (actual["gyroX"]-offsetGyroX) * scaleGyroX
    actual["gyroY"] = (actual["gyroY"]-offsetGyroY) * scaleGyroY
    actual["gyroZ"] = (actual["gyroZ"]-offsetGyroZ) * scaleGyroZ
    plt.figure(figsize=(20,10))
    plt.subplot(2,1,1)
    actual["sumAcc"] = (abs(actual["accX"]**2) + abs(actual["accY"]**2) + abs(actual["accZ"]**2))**0.5
    sns.lineplot(data=actual[["accX", "accY", "accZ", "sumAcc"]])
    plt.grid()
    plt.subplot(2,1,2)
    sns.lineplot(data=actual[["gyroX", "gyroY", "gyroZ"]])
    plt.grid()
```

```python
plot_data("FALLING/1.txt")
```

```python
plot_data("SITTING/1.txt")
```

```python
plot_data("RUNNING/2.txt")
```

```python
plot_data("WALKING/3.txt")
```

```python
plot_data("SITTINGDOWN/4.txt")
```

```python
plot_data("STANDINGUP/1.txt")
```
